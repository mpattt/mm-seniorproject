# Prepare Data : Calculate velocity

```
def getDistanceFromLatLonInKm(lat1,lon1,lat2,lon2):
    R = 6371 # Radius of the earth in km
    dLat = radians(lat2-lat1)
    dLon = radians(lon2-lon1)
    rLat1 = radians(lat1)
    rLat2 = radians(lat2)
    a = sin(dLat/2) * sin(dLat/2) + cos(rLat1) * cos(rLat2) * sin(dLon/2) * sin(dLon/2) 
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    d = R * c # Distance in km
    return d

def calc_velocity(dist_km, time_start, time_end):
    if time_end > time_start and float((time_end - time_start).seconds) != 0:
      speed=  dist_km / float((time_end - time_start).seconds)
    else :
      speed= 0
    return  speed

```

# Detect Stay point

```
#ฟังก์ชันการแปลงมุมเรเดียน func to convert radian
def rad(d):
    return float(d) * math.pi/180.0

#การคำนวณระยะทางระหว่างสองจุดตามละติจูดและลองจิจูดหน่วยเอาท์พุทคือหน่วยเมตร calculate distance between two lat long output unit as meter
EARTH_RADIUS=6378.137
def GetDistance(lng1,lat1,lng2,lat2):
    radLat1 = rad(lat1);
    radLat2 = rad(lat2);
    a = radLat1 - radLat2;
    b = rad(lng1) - rad(lng2);
    s = 2 * math.asin(math.sqrt(math.pow(math.sin(a/2),2) +math.cos(radLat1)*math.cos(radLat2)*math.pow(math.sin(b/2),2)))
    s = s * EARTH_RADIUS;
    s = round(s * 10000,2) / 10;
    return s

#คำนวณฟังก์ชั่นของจุดกึ่งกลางส่งคืนละติจูดและลองจิจูดของจุดกึ่งกลางข้อมูลคือชุดข้อมูลในรูปแบบ DataFrame, i, j คือตำแหน่ง, i <j
def ComputMeanCoord(data,i,j):
    latitude=list(data['Latitude'])
    longitude=list(data['Longitude'])
    amount=j-i+1
    if amount==0:
        return float(latitude[i]),float(longitude[i])
    sumLongitude=0
    sumLatitude=0
    for k in range(i,j+1):
        sumLatitude+=float(latitude[k])
        sumLongitude+=float(longitude[k])
    return sumLatitude/amount,sumLongitude/amount

#function to detect stay point
def StayPoint_Detection(data,distThreh,timeThreh):
    latitude=list(data['Latitude'])
    longitude=list(data['Longitude'])
    day=list(data['day']);hour=list(data['hour']);minute=list(data['minute']);second=list(data['second'])
    i=0
    SP=[]
    pointNumber=len(data) 
    while i<pointNumber:
        j=i+1
        Token=0
        while j<pointNumber:
            dist=GetDistance(longitude[i],latitude[i],longitude[j],latitude[j])
            if dist>float(distThreh):
                Time=abs(86400*(day[j-1]-day[i])+3600*(hour[j-1]-hour[i])+60*(minute[j-1]-minute[i])+(second[j-1]-second[i]))
                if Time>timeThreh:
                    S_latitude,S_longitude=ComputMeanCoord(data,i,j-1)
                    SarvT=str(year[i])+'-'+str(month[i])+'-'+str(day[i])+' '+str(hour[i])+':'+str(minute[i])+':'+str(second[i])
                    SlevT=str(year[j-1])+'-'+str(month[j-1])+'-'+str(day[j-1])+' '+str(hour[j-1])+':'+str(minute[j-1])+':'+str(second[j-1])
                    SP.append([S_longitude,S_latitude,SarvT,SlevT,[i,j-1]])
                    i=j
                    Token=1
                break
            j+=1
        if Token!=1:
            i+=1
    return SP
```

# Cluster Stay point by DBSCAN

```
def getDistanceFromLatLonInKm(lat1,lon1,lat2,lon2):
    R = 6371 # Radius of the earth in km
    dLat = radians(lat2-lat1)
    dLon = radians(lon2-lon1)
    rLat1 = radians(lat1)
    rLat2 = radians(lat2)
    a = sin(dLat/2) * sin(dLat/2) + cos(rLat1) * cos(rLat2) * sin(dLon/2) * sin(dLon/2) 
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    d = R * c # Distance in km
    return d

def haversine(lonlat1, lonlat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    
    """
    # convert decimal degrees to radians 
    lat1, lon1 = lonlat1
    lat2, lon2 = lonlat2
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    return c * r
```

```
#DBSCAN to cluster staypoint
for i in range(0,11):
  if i == 1:
    continue
  else:
    path='/poi/staypoint/staypoint'+str(i)+'.csv'
  data=pd.DataFrame()
  trajTem=pd.read_csv(path,sep=',')
  trajTem=trajTem.iloc[:, [2,1]]
  trajTem.columns=['lat','lon']
  data=data.append(trajTem,ignore_index=True)
  print(data.shape)
  distance_matrix = squareform(pdist(data, (lambda u,v: haversine(u,v))))
  
  db = DBSCAN(eps=0.1, min_samples=4, metric='precomputed',algorithm ='auto') 
  y_db = db.fit_predict(distance_matrix)

  data['cluster'] = y_db
  # Number of clusters in labels, ignoring noise if present.
  n_clusters_ = len(set(data['cluster'])) - (1 if -1 in data['cluster'] else 0)
  n_noise_ = list(data['cluster']).count(-1)

  print('Estimated number of clusters: %d' % n_clusters_)
  print('Estimated number of noise points: %d' % n_noise_)

  import statistics
  import gmplot
  latmin = statistics.median(data.lat)
  lonmax = statistics.median(data.lon)
  gmap = gmplot.GoogleMapPlotter(latmin,lonmax, 13)
  center=pd.DataFrame(columns=['user','lat','lon'])
  colors = ['red','blue','green','purple','orange','yellow','pink','white','red','blue','green','purple','orange','yellow','pink','white','red','blue','green','purple','orange','yellow','pink','white']
  for index in range(-1,n_clusters_-1):
    test = data[data['cluster'] == index]
    if index != -1:
      center = center.append({'user': str(i),'lat':statistics.mean(test.lat),'lon':statistics.mean(test.lon)}, ignore_index=True)
    gmap.scatter(test['lat'], test['lon'] ,color='#ffffff',size =20, marker = False)
  center.to_csv('/poi/staypoiny/center'+str(i)+'.csv')
  gmap.draw('/poi/map/mappoint'+str(i)+'.html') 
```
